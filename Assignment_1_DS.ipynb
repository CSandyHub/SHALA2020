You have a matrix $X$ of size (10,5) filled with random values. Print the following :

Let $Y$ denote a square matrix of size 10 with random values. Print the following:
* Determinant
* Rank of the matrix
* Diagonal Elements

* Eigen values

* Inverse if exists, otherwise Pseudo-Inverse

````
np.random.seed(3)
Y = np.random.rand(10,10)
print(Y)
print('Determinant = ',np.linalg.det(Y))
print('Rank = ',np.linalg.matrix_rank(Y))
print('Trace = ',Y.trace())
print('Eigen Values = ',np.linalg.eigvals(Y))
print('Matrices obtained after Singular Value Decomposition = ', np.linalg.svd(Y))
Y1=np.linalg.inv(Y)
print('Inverse = ',Y1)
print('Multiplication Y.X = ',np.linalg.multi_dot([Y,X]))
print('Multiplication Y and Y1 = ', Y.dot(Y1))
````

In this part, you will be working with the library ```pandas```. Import this library with an alias name ```pd```.

````
# Initial imports
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
````

Download dataset from: http://biostat.mc.vanderbilt.edu/wiki/pub/Main/DataSets/titanic3.csv
Read the data and display the first 5 rows

````
#Download dataset from: http://biostat.mc.vanderbilt.edu/wiki/pub/Main/DataSets/titanic3.csv
pd.set_option('display.max_columns', 20)
file_url = "http://biostat.mc.vanderbilt.edu/wiki/pub/Main/DataSets/titanic3.csv"
data=pd.read_csv(file_url)
data.head()

````

### Preprocessing of the Dataset



---


**1. Convert the categorical values of columns 'sex', 'embarked', 'pclass' into dummy values and add them into titanic_data_new DataFrame**




---



````
dmSex = pd.get_dummies(data.sex)
dmEmbarked=pd.get_dummies(data.embarked)
dmPclass= pd.get_dummies(data.pclass)
titanic_data_new= data.join((dmSex,dmEmbarked,dmPclass))
titanic_data_new
````



---


**2. Fill the missing data in Age and Fare columns with the mean value of the respected column then add these columns to titanic_data_new**


---



````
nonNullAge=data.age.replace(to_replace=np.nan,value=data.age.mean())
titanic_data_new['Modified_Age']=nonNullAge
nonNullFare=data.age.replace(to_replace=np.nan,value=data.fare.mean())
titanic_data_new['Modified_Fare']=nonNullFare
titanic_data_new
````

###Normalization


---



1. Normalize the columns which have numerical data



---



````
numericColumnsDf=titanic_data_new.select_dtypes(include=np.number)
normalizedDf=numericColumnsDf -numericColumnsDf.mean() / numericColumnsDf.std()
normalizedDf
````



---
2. Drop the original age, fare columns
3. Append the normalized columns to the titanic_data_new DataFrame.



---



````
titanic_data_new.drop(['sex','age'], axis=1,inplace=True)

````

````
titanic_data_new=titanic_data_new.join(normalizedDf,rsuffix='_normalized')
titanic_data_new
````

###Plotting
1. plot the age against the ID(continuous)
2. plot the fare against the ID(scatter)

````
import matplotlib.pyplot as plt

plt.plot(titanic_data_new.index,titanic_data_new['age'])
plt.show()
plt.scatter(titanic_data_new.index,titanic_data_new['fare'])
plt.show()
````

Plot $\sin(x)$ and its spectrum, $\cos(x)$ and its spectrum in a single plot, in the domain $[-2\pi,2\pi]$. Label the plot appropriately.

````
x=np.linspace(-2*np.pi,2*np.pi,300)
plt.figure(figsize=(10,10))
y1 = np.sin(x)
ffty1=np.fft.fftn(y1)
y2 = np.cos(x)
ffty2=np.fft.ifftn(y2)
plt.plot(x, y1,'-b',label="$\sin(x)$")
plt.plot(x, np.real(ffty1),'-r',label="FFT $\sin(x)$")
plt.plot(x, y2,'-g',label="$\cos(x)$")
plt.plot(x, np.real(ffty2),'-y',label="FFT $\cos(x)$")
plt.legend(loc="lower left")
plt.show()

````
